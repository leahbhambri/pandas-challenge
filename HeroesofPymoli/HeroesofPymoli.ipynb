```python
#import dependancies
import pandas as pd
import numpy as np
```

```python
#define file path and read csv
file = "Resources/purchase_data.csv"
df=pd.read_csv(file)
df.head()
```

```python
#define total players count
players = []
for SN in df["SN"]:
    if SN not in players:
        players.append(SN)
player_count = len(players)
#define total unique items
unique_list = []

for item in df["Item Name"]:
    if item not in unique_list:
        unique_list.append(item)
            
#define average price
average_price = df['Price'].mean()

#define total purchases
total_purchases = df['SN'].count()

#define revenue
revenue = df["Price"].sum()
```

```python
player_count
```

```python
#print summary dataframe
summary_df = pd.DataFrame({
    "Number of Unique Items" : [len(unique_list)],
    "Average Price" : [average_price],
    "Number of Purchase" : [total_purchases],
    "Total Revenue" : [revenue]
})
summary_df['Average Price'] = summary_df['Average Price'].map("${:.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:,.2f}".format)
summary_df
```

```python
#drop duplicates of SN
gender_df = df[["SN", "Gender",]].drop_duplicates()

#create df grouped by gender
gender_count = gender_df["Gender"].value_counts()
```

```python
#create gender dataframe grouped by gender
gender_count_df = gender_df.groupby(["Gender"])
```

```python
#define total players and percentages
total_players = len(players)
percentage_players = gender_count/total_players*100
```

```python
#create summary dataframe
summary_gender_df= pd.DataFrame({
    "Total Count" : gender_count,
    "Percentage of Players" : percentage_players
})
summary_gender_df['Percentage of Players']= summary_gender_df['Percentage of Players'].map("{:.2f}%".format)

summary_gender_df
```

```python
#groupby dataframe will all purchases
gender_count_total = df["Gender"].value_counts()
grouped_gender_df=df.groupby(["Gender"])
```

```python
#create variables for summary table
average_price_gender = grouped_gender_df["Price"].mean()
total_price = grouped_gender_df["Price"].sum()
purchase_person = total_price/gender_count
```

```python
#create dataframe
purchasing_gender_df= pd.DataFrame({
    "Purchase Count" : gender_count_total,
    "Average Purchase Price" : average_price_gender,
    "Total Purchase Value" : total_price,
    "Avg Total Purchase per Person" : purchase_person  
})
purchasing_gender_df['Average Purchase Price'] = purchasing_gender_df['Average Purchase Price'].map("${:,.2f}".format)
purchasing_gender_df['Total Purchase Value'] = purchasing_gender_df['Total Purchase Value'].map("${:,.2f}".format)
purchasing_gender_df['Avg Total Purchase per Person'] = purchasing_gender_df['Avg Total Purchase per Person'].map("${:,.2f}".format)
purchasing_gender_df.head()
```

```python
#create age bins
bins = [0,9,14,19,24,29,34,39,110]
group_names = ("<10","10-14","15-19","20-24","25-29", "30-34","35-39","40+")
df["Age Bucket"] = pd.cut(df["Age"], bins, labels = group_names, include_lowest = True)
```

```python
#remove duplicates of SN from dataframe
age_df = df.drop_duplicates(subset=['SN'])
```

```python
#get age bucket counts
age_group=age_df['Age Bucket'].value_counts()
```

```python
#print dataframe image
age_group_df = age_df.groupby(['Age Bucket'])
```

```python
#create total and percentage values
percentage_age=age_group/len(players)
summary_age_df= pd.DataFrame({
    "Total Count" : age_group,
    "Percentage of Players" : percentage_age
})
summary_age_df['Percentage of Players'] = summary_age_df['Percentage of Players'].map("{:.2%}".format)

summary_age_df
```

```python
#create df group with all purchases
grouped_age_df=df.groupby(["Age Bucket"])
```

```python
#create summary table variables
purchase_count = grouped_age_df["Age Bucket"].count()
average_age_purchase = grouped_age_df['Price'].mean()
total_age_purchase = grouped_age_df['Price'].sum()
avg_total_person = total_age_purchase/age_group
```

```python
#create summary dataframe
purchasing_age_df= pd.DataFrame({
    "Purchase Count" : purchase_count,
    "Average Purchase Price" : average_age_purchase,
    "Total Purchase Value" : total_age_purchase,
    "Avg Total Purchase per Person" : avg_total_person    
})
purchasing_age_df['Average Purchase Price'] = purchasing_age_df['Average Purchase Price'].map("${:,.2f}".format)
purchasing_age_df['Total Purchase Value'] = purchasing_age_df['Total Purchase Value'].map("${:,.2f}".format)
purchasing_age_df['Avg Total Purchase per Person'] = purchasing_age_df['Avg Total Purchase per Person'].map("${:,.2f}".format)
purchasing_age_df
```

```python
person_stats = df['SN'].value_counts()
person_stats_df = df.groupby(['SN'])
```

```python
sn_count = person_stats_df['SN'].count()
sn_avg_price = person_stats_df['Price'].mean()
sn_total_price = person_stats_df['Price'].sum()
```

```python
sn_df = pd.DataFrame({
    "Purchase Count" : sn_count,
    "Average Purchase Price" : sn_avg_price,
    "Total Purchase Price" : sn_total_price    
})

sn_sorted_df = sn_df.sort_values("Total Purchase Price", ascending = False)
sn_sorted_df['Average Purchase Price'] = sn_sorted_df['Average Purchase Price'].map("${:,.2f}".format)
sn_sorted_df['Total Purchase Price'] = sn_sorted_df['Total Purchase Price'].map("${:,.2f}".format)
sn_sorted_df.head()
```
